```python
import xlsxwriter
workbook = xlsxwriter.Workbook('4.310VDC_Pump7V.xlsx')
worksheet = workbook.add_worksheet()
# Start from the first cell. Rows and columns are zero indexed.
row = 0
col = 0

# Iterate over the data and write it out row by row.
for Time, Inlet, Outlet, Bucket,InBucket,OutBucket,Atmosphere in (data1):
    worksheet.write(row, col,    Time )
    worksheet.write(row, col + 1, Inlet)
    worksheet.write(row, col + 2, Outlet)
    worksheet.write(row, col + 3, InBucket)
    worksheet.write(row, col + 4, OutBucket)
    worksheet.write(row, col + 5, Atmosphere)

    row += 1
    
workbook.close()
```
